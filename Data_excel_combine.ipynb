```python
import pandas as pd  
import numpy as np  
import glob  
import sys

#파일 Union  
all_data = pd.DataFrame()  
for f in glob.glob('/Users/jeonmyeong-un/Desktop/KDT LIKELION/KDT_Project/data/2019/melon_2019.*.xlsx'): # 예를들어 201901, 201902 로 된 파일이면 2019_*  
    df = pd.read_excel(f)  
    all_data = all_data.append(df, ignore_index=True)

te = pd.DataFrame(all_data)
te = te.drop('Unnamed: 0', axis=1)
te


#데이터 잘 들어오는지 확인  
# all_data.head()

# 2020 전체 데이터 파일저장  
te.to_excel("/Users/jeonmyeong-un/Desktop/KDT LIKELION/KDT_Project/data/melon_2019.xlsx", header=True, index=False)
```
